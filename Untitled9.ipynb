```python
import numpy as np

# 1. Create 4x4 random array
arr = np.random.randint(10, 100, (4,4))
print("Array:\n", arr)

# 2. Third row and second column
print("Third row:", arr[2])
print("Second column:", arr[:,1])

# 3. Sum of each row
print("Row sums:", arr.sum(axis=1))

# 4. Maximum value in each column
print("Column max:", arr.max(axis=0))

# 5. Replace values > 50 with 50
arr[arr > 50] = 50
print("Modified array:\n", arr)
```
